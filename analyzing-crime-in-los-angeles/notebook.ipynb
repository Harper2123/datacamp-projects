## Task 1
Which hour has the highest frequency of crimes? Store as an integer variable called `peak_crime_hour`.

```python
crimes.loc[:,"hour_occurence"] = crimes.loc[:, "TIME OCC"]
crimes["hour_occurence"] = crimes["hour_occurence"].astype(int)
crimes["exact_hour_occ"] = crimes["hour_occurence"] // 100
peak_crime_hour = crimes["exact_hour_occ"].value_counts().idxmax()
```

## Task 2
Which area has the largest frequency of night crimes (crimes committed between 10 pm and 3:59 am)? Save as a string variable called `peak_night_crime_location`.

```python
mask = ((crimes["hour_occurence"] >= 2200) | ((crimes["hour_occurence"] >= 0) & (crimes["hour_occurence"] <= 359)))

crimes_1 = crimes[mask]
peak_night_crime_location = crimes_1["AREA NAME"].value_counts().idxmax()
```

## Task 3
Identify the number of crimes committed against victims of different age groups. Save as a pandas Series called `victim_ages`, with age group labels `0-17`, `18-25`, `26-34`, `35-44`, `45-54`, `54-64`, and `65+` as the index and the frequency of crimes as the values.

```python
bins = [0, 17, 25, 34, 44, 54, 64, float('inf')]
labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

age_groups = pd.cut(crimes["Vict Age"], bins=bins, labels=labels, right=True, include_lowest=True)

victim_ages = age_groups.value_counts().sort_index()
print(victim_ages)
```
