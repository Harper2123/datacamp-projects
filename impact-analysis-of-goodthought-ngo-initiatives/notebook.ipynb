## Task 1

List the top five assignments based on total value of donations, categorized by donor type. The output should include four columns:
1. `assignment_name`
2. `region`
3. `rounded_total_donation_amount` rounded to two decimal places
4. `donor_type`

Sort the results by `rounded_total_donation_amount` in **descending** order.

```python
-- highest_donation_assignments
WITH donation_total AS (
	SELECT
		donations.assignment_id,
		donors.donor_type,
		SUM(donations.amount) AS total_amount
	FROM
		donations
	LEFT JOIN
		donors
	ON
		donations.donor_id = donors.donor_id
	GROUP BY
		donations.assignment_id,
		donors.donor_type
),
highest_donation_assignments AS (
	SELECT
		assignments.assignment_name,
		assignments.region,
		ROUND(SUM(donation_total.total_amount), 2) AS rounded_total_donation_amount,
		donation_total.donor_type
	FROM
		donation_total
	LEFT JOIN
		assignments
	ON
		assignments.assignment_id = donation_total.assignment_id
	GROUP BY
		assignments.assignment_id,
		assignments.assignment_name,
		assignments.region,
		donation_total.donor_type
)
SELECT *
FROM
	highest_donation_assignments
ORDER BY rounded_total_donation_amount DESC
LIMIT 5;
```

## Task 2

Identify the assignment with the highest impact score in each region, ensuring that each listed assignment has received at least one donation. The output should include four columns:
1. `assignment_name`
2. `region`
3. `impact_score`
4. `num_total_donations`

Sort the results by `region` in **ascending order.** Include only the highest-scoring assignment per region, avoiding duplicates within the same region.

```python
-- top_regional_impact_assignments
WITH top_reg_impact AS (
	SELECT
		assignments.assignment_name,
		assignments.region,
		assignments.impact_score,
		SUM(donations.amount) AS num_total_donations,
		ROW_NUMBER() OVER(PARTITION BY assignments.region ORDER BY assignments.impact_score DESC) AS region_rank
	FROM
		assignments
	LEFT JOIN
		donations
	ON
		assignments.assignment_id = donations.assignment_id
	GROUP BY
		assignments.assignment_id,
		assignments.assignment_name,
		assignments.region,
		assignments.impact_score
)
SELECT
	assignment_name,
	region,
	impact_score,
	num_total_donations
FROM
	top_reg_impact
WHERE
	region_rank = 1
ORDER BY region;
```
